```python
    from elasticutils import get_es, S
import pandas as pd

URL = "http://localhost:9200"
INDEX = "logstash-2014.07.06"
_es = S().es(urls=[URL]).indexes(INDEX)

data_list = []
_es.start = 0
_es.stop = 100000
for item in _es.execute():
    data_list.append(item._results_dict)

df = pd.DataFrame(data_list)
```

```python
def geturl(s):
    return s.split('?')[0]
df['menu'] = df['url'].map(geturl)

def istarget(s):
    return not(s.startswith('/api') or s.startswith('/stat'))
df['target'] = df['menu'].map(istarget)

df2 = df[df['target']]
```
