```python
import pgdb
import pandas as pd
import numpy as np
from pandas.io import sql
db = pgdb.connect(database="soma",host="기업정보입니다",user="기업정보입니다",password="기업정보입니다")
users = sql.read_frame("select * from users",db)
```

```python
df_jubu = pd.read_csv("user_profiled_jubu.log")
df_not_jubu = pd.read_csv("temp.log")

df2_jubu = df_jubu.merge(users,left_on='xid', right_on='user_id', how='inner')
df2_not_jubu = df_not_jubu.merge(users,left_on='xid', right_on='user_id', how='inner')
```

```python
pos_review = []
for user in df2_jubu['xid'].unique():
    df_user_data = df2_jubu[df2_jubu['xid']==user]
    string_data = ''
    for i in range(len(df_user_data)):
        curr_string = str(df_user_data.iloc[i]['cate'])
        if curr_string != 'nan':
            string_data = string_data + curr_string + ' '
            
        curr_string = str(df_user_data.iloc[i]['query'])
        if curr_string != 'nan':
            string_data = string_data + curr_string + ' '
    pos_review.append(string_data)

neg_review = []
for user in df2_not_jubu['xid'].unique():
    df_user_data = df2_not_jubu[df2_not_jubu['xid']==user]
    string_data = ''
    for i in range(len(df_user_data)):
        curr_string = str(df_user_data.iloc[i]['cate'])
        if curr_string != 'nan':
            string_data = string_data + curr_string + ' '
            
        curr_string = str(df_user_data.iloc[i]['query'])
        if curr_string != 'nan': 
            string_data = string_data + curr_string + ' '
    neg_review.append(string_data)
```

```python
from gensim import corpora
pos_texts = [[word for word in list(document.decode("utf-8"))] for document in pos_review]
neg_texts = [[word for word in list(document.decode("utf-8"))] for document in neg_review]
all_texts = pos_texts + neg_texts
dictioanry = corpora.Dictionary(all_texts)
pos_vects = [dictioanry.doc2bow(text) for text in pos_texts]
neg_vects = [dictioanry.doc2bow(text) for text in neg_texts]
```

```python
from sklearn import cross_validation
from sklearn.svm import LinearSVC
import numpy as np
X = pos_vects + neg_vects
y = [1]*len(pos_vects)+[2]*len(neg_vects)
corpora.svmlightcorpus.SvmLightCorpus.save_corpus('polarity.svmlight',X,labels=y)
from sklearn.datasets import load_svmlight_file
import joblib
X, y = load_svmlight_file("polarity.svmlight")
X_train, X_test, y_train, y_test = cross_validation.train_test_split(X, y, test_size=0.2, random_state=0)
clf = LinearSVC(loss='l2', C=0.1).fit(X_train, y_train)
print clf.score(X_test, y_test)
joblib.dump(clf,'polarity.svm.model',compress=3)
```
