```python
sorted_df = newdf.sort(['xid', 'ymd'])
sorted_df[0:20]
```

```python
time_diff = dict()
for i in range(1, len(sorted_df)):
    if sorted_df.iloc[i].xid == sorted_df.iloc[i-1].xid:
        diff = int(sorted_df.iloc[i].ymd) - int(sorted_df.iloc[i-1].ymd)
        if diff not in time_diff: time_diff[diff] = 0
        time_diff[diff] += 1
for key, value in time_diff.iteritems():
    print key, value
    if key > 1000: break
```

```python
def getage(s):
    if s == 0:
        return 0
    return 2014 - int(s / 10000) + 1
sorted_df['age'] = sorted_df['birth'].map(getage)
sorted_df[50:70]
```

```python
time_threshold = 300
pos_texts = []
neg_texts = []
curr_text = []
is_pos = False
for i in range(0, len(sorted_df)):
    if (i == 0) or (sorted_df.iloc[i].xid != sorted_df.iloc[i-1].xid) or (int(sorted_df.iloc[i].ymd) - int(sorted_df.iloc[i-1].ymd) > time_threshold):
        if len(curr_text) > 0:
            if is_pos:
                pos_texts.append(curr_text)
            else:
                neg_texts.append(curr_text)
        curr_text = []
    curr_str = unicode(str(sorted_df.iloc[i].query).replace(' ', ''), 'utf-8')
    for j in range(0, len(curr_str)-1):
        curr_text.append(curr_str[j:j+2])
    is_pos = (sorted_df.iloc[i].gender == 'male')

if len(curr_text) > 0:
    if is_pos:
        pos_texts.append(curr_text)
    else:
        neg_texts.append(curr_text)
```

```python
print len(pos_texts)
print len(neg_texts)
print float(len(pos_texts)) / float(len(pos_texts) + len(neg_texts))
```

```python
from gensim import corpora
all_texts = pos_texts + neg_texts
dictioanry = corpora.Dictionary(all_texts)
pos_vects = [dictioanry.doc2bow(text) for text in pos_texts]
neg_vects = [dictioanry.doc2bow(text) for text in neg_texts]

from sklearn import cross_validation
from sklearn.svm import LinearSVC
import numpy as np
neg_vects2 = neg_vects[:len(pos_vects)]
X = pos_vects + neg_vects2
y = [1]*len(pos_vects)+[2]*len(neg_vects2)
corpora.svmlightcorpus.SvmLightCorpus.save_corpus('polarity.svmlight',X,labels=y)
from sklearn.datasets import load_svmlight_file
import joblib
X, y = load_svmlight_file("polarity.svmlight")
X_train, X_test, y_train, y_test = cross_validation.train_test_split(X, y, test_size=0.2, random_state=0)
clf = LinearSVC(loss='l2', C=0.1).fit(X_train, y_train)
print clf.score(X_test, y_test)
joblib.dump(clf,'polarity.svm.model',compress=3)
```
